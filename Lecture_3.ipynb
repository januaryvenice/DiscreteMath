The Graph ***G*** consists of three vertices {A, B, C} and two edges {(A,B),(C,C)}.

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 3))
graph = nx.Graph()
graph.add_edge("A", "B")
graph.add_node("C")
graph.add_edge("C", "C")
nx.draw_spring(graph, with_labels = True)
plt.title("G")
plt.show()
```

The Directed Graph (Digraphs) ***G*** consists of the vertex set {A,B,C,D} and the ordered edge set {(A,B),(B,C),(C,D),(D,A)}.

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 3))
digra = nx.DiGraph()
digra.add_edge("A", "B")
digra.add_edge("B", "C")
digra.add_edge("C", "D")
digra.add_edge("D", "A")
nx.draw(digra, with_labels = True)
plt.title("G")
plt.show()
```

The Bipartite Graph **G** consists of the setA with vertices {1,2,3,4} and setB with vertices {5,6,7}.

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
bigra = nx.Graph()
setA = [1, 2, 3, 4]
setB = [5, 6, 7]
bigra.add_nodes_from(setA, bipartite=0)
bigra.add_nodes_from(setB, bipartite=1)
for nodeA in setA:
    for nodeB in setB:
        bigra.add_edge(nodeA, nodeB)
pos = {node: (0, i) for i, node in enumerate(setA)}
pos.update({node: (1, i) for i, node in enumerate(setB)})
nx.draw(bigra, pos, node_size=500, with_labels=True, width=1.5)
plt.title("G")
plt.show()
```

Let v be the vertex. The degrees of the vertices in Graph **G** is the number of edges incident to the v.

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "a", 1: "b", 2: "c", 3: "d", 4: "e"})
exgra = nx.from_edgelist([(4, "X"), ("X", "a")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("G")
plt.show()

print("deg(v): ")
print(dict(G.degree))

```

The Handshake theorem states that in any undirected graph, the sum of the degrees of all vertices is equal to twice the number of edges in a graph.

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "a", 1: "b", 2: "c", 3: "d", 4: "e"})
exgra = nx.from_edgelist([(4, "X"), ("X", "a")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("G")
plt.show()

print("Edges * 2: ")
print(nx.number_of_edges(G) * 2)
degrees = G.degree()
result = sum(deg for node, deg in degrees)
print("Sum of degrees of all vertices: ")
print(result)
print("Therefore, the theorem is true.")
```

Travelling in the graph G can be done through a walk, a trail, or a path.:

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "a", 1: "b", 2: "c", 3: "d", 4: "e"})
exgra = nx.from_edgelist([(4, "X"), ("X", "a")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("G")
plt.show()
```


*   **Walk** from a→1: a → x → 4 →1
*   **Trail** is a walk with no repeated edge. An example of a trail from a→ 0: a→ x→ 4→ 2→ 0
*   **Path** is a walk with no repeated vertex. An example of a path from a→ 2: a→ x→ 4→ 1→ 3→ 0→ 2
*   **Closed Walk** is a walk that has the same start and end vertex. A closed walk of a is a→ c→ b→ d→ a
*   **Circuit** is a closed trail. An example of a circuit of 1: 1→ 2→ 3→ 0→ 2→ 4→ 1
*   **Simple Circuit** is a circuit with only the start and end vertices as equal. An example of a simple circuit of 1: 1→ 2→ 0→ 1

Euler's Trail for Graph **G**: a → f → c → h → d → c → g → d → f → g → b → f → e → a

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
G = nx.Graph()
edges = [("a", "e"), ("e", "f"), ("f", "b"), ("b", "g"), ("g", "f"), ("f", "c"), ("c", "d"), ("d", "h"), ("h", "c"), ("c", "g"), ("g", "d"), ("d", "f"), ("f", "a")]
G.add_edges_from(edges)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, font_size=10)
plt.title("G")
plt.show()
```

**The Travelling Salesman Problem**

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()
G.add_edge("A", "B", weight=4)
G.add_edge("B", "C", weight=8)
G.add_edge("A", "C", weight=10)
G.add_edge("A", "D", weight=6)
G.add_edge("D", "E", weight=9)
G.add_edge("E", "F", weight=1)
G.add_edge("B", "F", weight=5)
G.add_edge("C", "F", weight=7)
G.add_edge("C", "E", weight=2)
G.add_edge("F", "A", weight=3)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```

The best route is A → B → F → C → E → D → A
