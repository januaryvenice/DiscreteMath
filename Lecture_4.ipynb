**Main Graph**

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
# Use seed for reproducibility
G = nx.Graph()
plt.figure(figsize=(10, 8))
# Add nodes
G.add_nodes_from(["A","B","C","D","E","F","G"])
# Add self-loops to the remaining nodes
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
nx.draw(G, pos=pos,with_labels=True)
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True)

plt.show()
```

**Kruskal's Algorithm**

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])
edgelist = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("B", "E"), ("I", "E"), ("E", "F"), ("F", "G"), ("G", "E")]
G.add_edges_from(edgelist)

plt.figure(figsize=(10, 8))
K = nx.Graph()
K.add_nodes_from(G.nodes)
edges = [("A", "I"), ("A", "C"), ("B", "C"), ("C", "D"), ("C", "E"), ("E", "F"), ("E", "G")]
K.add_edges_from(edges)
pos = nx.planar_layout(G)
nx.draw_networkx_edge_labels(
    K, pos,
    edge_labels={("A", "I"):5, ("A", "C"):10, ("B", "C"):20, ("C", "D"):30, ("C", "E"):15, ("E", "F"):35, ("E", "G"):45},
    font_color='purple'
)
nx.draw(
    K, pos=pos, node_color='orange',
    with_labels=True
)

plt.show()
```

| Iteration | Edge      | Weight | Action     |
|-----------|-----------|--------|------------|
| 1         | ('A', 'I')| 5      | Add        |

| 3         | ('C', 'E')| 15     | Add        |
| 4         | ('C', 'B')| 20     | Add        |
| 5         | ('B', 'E')| 25     | Not Added  |
| 6         | ('C', 'D')| 30     | Add        |
| 7         | ('E', 'F')| 35     | Add        |
| 8         | ('I', 'E')| 40     | Not Added  |
| 9         | ('E', 'G')| 45     | Add        |

**Prim's Algorithm**

```python
import matplotlib.pyplot as plt
import networkx as nx

seed = 55
G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])
edgelist = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("B", "E"), ("I", "E"), ("E", "F"), ("F", "G"), ("G", "E")]
G.add_edges_from(edgelist)

plt.figure(figsize=(10,8))
P = nx.Graph()
P.add_nodes_from(G.nodes)
primdedge = [("A", "I"), ("A", "C"), ("C", "B"), ("C", "D"), ("C", "E"), ("E", "F"), ("E", "G")]
P.add_edges_from(primdedge)
pos = nx.planar_layout(G)
nx.draw_networkx_edge_labels(
    P, pos,
    edge_labels={("A", "I"):5, ("A", "C"):10, ("B", "C"):20, ("C", "D"):30, ("C", "E"):15, ("E", "F"):35, ("E", "G"):45},
    font_color='purple'
)
nx.draw(
    P, pos=pos, node_color='orange',
    with_labels=True
)

plt.show()
```

| Iteration | Vertex | Edge      | Weight |
|-----------|--------|-----------|--------|
| 0         | I      |           |        |

| 2         | C      | ('A', 'C')| 10     |
| 3         | E      | ('C', 'E')| 15     |
| 4         | B      | ('C', 'B')| 20     |
| 5         | D      | ('C', 'D')| 30     |

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])
edge_list = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("B", "E"), ("I", "E"), ("E", "F"), ("F", "G"), ("G", "E")]
G.add_edges_from(edge_list)

plt.figure(figsize=(10,8))
D = nx.Graph()
D.add_nodes_from(G.nodes)
D.add_edges_from(G.edges)
pos = nx.planar_layout(G)
nx.draw_networkx_edge_labels(
    D, pos,
    edge_labels={("A", "C"):10, ("A", "I"):5, ("C", "B"):20, ("C", "E"):15, ("C", "D"):30, ("B", "E"):25, ("I", "E"):40, ("E", "F"):35, ("F", "G"):50, ("G", "E"):45},
    font_color='purple'
)

nx.draw_networkx_edges(
    D, pos=pos,
    edgelist=[("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("G", "E"), ("E", "F")],
    edge_color='red'
)
nx.draw(
    D, pos=pos, node_color='orange',
    with_labels=True
)

plt.show()
```
